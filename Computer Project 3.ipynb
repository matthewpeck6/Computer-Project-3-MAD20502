```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Question 1

```python
D = {(1,3),(2,1),(3,2),(4,0)} #given data set
D = list(D)
n = len(D)

#summation of a x and y values
#calculation of the mean of y (y bar)
#defined beforehand so that variables can be called on in all parts
sum_x = 0
sum_y = 0
for i in range(n):
    r = D[i]
    sum_x = sum_x + r[0]
    sum_y = sum_y + r[1]

y_bar = sum_y/n
```

### Part A and B

```python
plt.figure(figsize = (8,6))
x = np.linspace(0, 5, 5)

#plots the points in the given data set
for i in range(n):
    r = D[i]
    plt.scatter(r[0], r[1])
    
aa = (sum_y - n)/sum_x    #a value for form a
ab = (sum_y - 2*sum_x)/n  #a value for form b

plt.plot(x, aa*x + 1, 'r-', label='form y = ax + 1') #graphing form a
plt.plot(x, 2*x + ab, label='form y = 2x + a')       #graphing form b
plt.plot(x, x*0 + y_bar, 'y--', label='mean of y')   #graphing mean of y

plt.xlim(0,5)
plt.ylim(-1,4)
plt.title(f'Plot of Data Set D and Two Best Fit Curves')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc = 0)
plt.show()
```

### Part C

```python
SSRESa = 0
SSREGa = 0
for i in range(n):
    #finding SSRES and SSREG of form a
    r = D[i]
    SSRESa = SSRESa + (r[1] - aa*r[0] - 1)**2
    SSREGa = SSREGa + (aa*r[0] + 1 - y_bar)**2

SSTOTa = SSRESa + SSREGa
Rsqa = SSREGa / SSTOTa
```

```python
SSRESb = 0
SSREGb = 0
for i in range(n):
    #finding SSRES and SSREG of form b
    r = D[i]
    SSRESb = SSRESb + (r[1] - 2*r[0] - ab)**2
    SSREGb = SSREGb + (2*r[0] + ab - y_bar)**2

SSTOTb = SSRESb + SSREGb
Rsqb = SSREGb / SSTOTb
```

```python
print(f'R^2 of form y = ax + 1 (form A) : {Rsqa}')
print(f'R^2 of form y = 2x + a (form B) : {Rsqb}')
```

The $R^2$ value of form B is greater than the $R^2$ value of form A:
<br> $0 < 0.02857 < 0.32787 < 1$
<br> This means that the <b>model of form B</b> fits the data better than the model of form A.

## Question 2

```python
#creating an array from 0 to 1000, then multiplying it by 2pi/1000 to get 0 to 2pi in 1000 increments
plt.figure(figsize = (8,6))
numArray = np.array(range(0, 1001))
myInt = (2 * np.pi) / 1000
newArray = numArray * myInt

#for loop adding the output from the x and y formulas to two arrays
xArray = []
yArray = []
for x in newArray:
    xInt = 3 * (x - np.sin(x))
    yInt = 3 * (1 - np.cos(x))
    xArray.append(xInt)
    yArray.append(yInt)

#plotting the two arrays
plt.plot(xArray, yArray)
plt.title('Cycloid Plot of 0 to 2π')
plt.xlabel('X axis', fontsize = 18)
plt.ylabel('Y axis', fontsize = 18)
plt.xlim(0, 19)
plt.ylim(0, 6.5)
plt.grid()
plt.show()
```

## Question 3

```python
def triforce_sierpinkski(num_points):
  
    #2 lists: 1 for x-coor and 1 for y_coor
    plt.figure(figsize = (8,6))
    x = [0.0]
    y = [0.0]

    #Gets the next point using the prev points coordinate 
    #and number passed in from random number generator
    def next_point(last_X_coor, last_Y_coor, num):

        if (num==1):
        #Corner 1 Calculation
            x_coor = last_X_coor / 2.0
            y_coor = last_Y_coor / 2.0
            x.insert(len(x), x_coor)
            y.insert(len(y), y_coor)
        elif (num==2):
        #Corner 2 Calculation
            x_coor = (last_X_coor + 1.0) / 2.0
            y_coor = last_Y_coor / 2.0
            x.insert(len(x), x_coor)
            y.insert(len(y), y_coor)
        elif (num==3):
        #Corner 3 Calculation
            x_coor = (last_X_coor + 0.5)/2.0
            y_coor = (last_Y_coor + np.sqrt(2.0))/2.0
            x.insert(len(x), x_coor)
            y.insert(len(y), y_coor) 
        else:
        #This should never happen
            print("How did it even get this number?")

    #Generates points and the specific corner
    #is based on random number between 1 and 3
    i = 0
    while i < num_points:
        position = random.randint(1,3)
        i += 1
        next_point(x[i-1], y[i-1], position)

    #Plots graph
    plt.plot(x, y, 'b.', markersize = .1)
    plt.imshow
```

```python
triforce_sierpinkski(100000)
```
